```python
df_can = pd.read_excel('./Canda.xlsx',
                      sheet_name='Canada by Citizenship',
                      skiprows=range(20),
                      skipfooter=2)
print('Data downloaded and read info a data frame')
```

```python
df_can.head()
```

```python
df_can.shape
```

```python
df_can = df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis=1)
```

```python
df_can.head()
```

```python
df_can = df_can.rename(columns={
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region'
})
df_can.head()
```

```python
all(isinstance(column, str) for column in df_can.columns)
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
df_can = df_can.set_index('Country')
```

```python
df_can.head()
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.head()
```

```python
df_can.shape
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
years
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
mpl.__version__
```

```python
df_top15.plot(kind='barh',
             figsize=(12, 12),
             color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntricies Contributing to the Immigration to Canada between 1980 - 2013')

for index, value in enumerate(df_top15):
    label = format(int(value), ',')
    plt.annotate(label,
                xy=(value - 47000, index - 0.10), 
                color='white')
plt.show()
```
